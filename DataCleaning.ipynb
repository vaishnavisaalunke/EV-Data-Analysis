Q5. How can Vehicle Location (GPS coordinates) be cleaned or converted for better readability?
---GPS coordinates are stored as text in POINT(long lat) format. code extracts longitude and latitude into separate numeric columns coordinates enable geographical visualization.

```python
df[['Longitude','Latitude']] = df['Vehicle Location'] \
    .str.replace('POINT \\(|\\)', '', regex=True) \
    .str.split(' ', expand=True)

df[['Longitude','Latitude']] = df[['Longitude','Latitude']].astype(float)
```
