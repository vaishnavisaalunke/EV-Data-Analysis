```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df.head()
```

What are the top 5 most common EV makes and models in the dataset?


```python
# Top 5 EV Makes
top_makes = df['Make'].value_counts().head(5)
top_makes
```

```python
# Top 5 EV Models
top_models = df['Model'].value_counts().head(5)
top_models
```

---It Counts the number of vehicles per manufacturer and Helps identify the most popular EV brands.Tesla is expected to dominate the dataset.

What is the distribution of EVs by county? Which county has the most registrations?

```python
county_distribution = df['County'].value_counts()
county_distribution.head(10)
```

---code counts EV registrations in each county.
King County has the highest number of EV registrations.

How has EV adoption changed over different model years?

```python
ev_by_year = df.groupby('Model Year').size()
ev_by_year
```

--- Vehicles are grouped by their manufacturing year and The results show a steady increase in EV adoption, especially after 2018.

What is the average electric range of EVs in the dataset?

```python
average_range = df['Electric Range'].mean()
average_range
```

---This calculates the mean electric range of all EVs and Ihe average electric range is approximately 200â€“230 miles.Newer EVs tend to have better battery efficiency.

What percentage of EVs are eligible for Clean Alternative Fuel Vehicle (CAFV) incentives?

```python
cafv_percentage = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] \
    .value_counts(normalize=True) * 100

cafv_percentage
```

---normalize=True converts counts into percentages. Most EVs are CAFV eligible, meaning they qualify for clean fuel incentives.Some vehicles are not eligible or have unknown eligibility due to missing range data.

How does the electric range vary across different makes and models?

```python
range_by_make = df.groupby('Make')['Electric Range'].mean().sort_values(ascending=False)
range_by_make.head(10)
```

```python
range_by_model = df.groupby('Model')['Electric Range'].mean().sort_values(ascending=False)
range_by_model.head(10)
```

--- Tesla models have the highest average electric range.Older models like early Nissan Leaf have lower range.

What is the average Base MSRP for each EV model?

```python
avg_msrp_model = df.groupby('Model')['Base MSRP'].mean().sort_values(ascending=False)
avg_msrp_model.head(10)
```

Are there any regional trends in EV adoption (e.g., urban vs. rural areas)?

```python
city_distribution = df['City'].value_counts()
city_distribution.head(10)
```
